```python
import pandas as pd
import numpy as np
df = pd.read_csv('train.csv',skipinitialspace=True)
dftest = pd.read_csv('test.csv',skipinitialspace = True)
```

```python
import spacy
from spacy.lang.en import English
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
nlp_eng=English()
stopwords=list(STOP_WORDS)
```

```python
# 建立SpaCy預處理function #Lemmatization #Remove Stopword & Punct
def tokenizer(sentence):
    mytokens=nlp_eng(sentence)
    mytokens=[word.lemma_.lower().strip() for word in mytokens if word.is_stop==False and word.is_punct==False]
    return mytokens
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix
```

```python
# 建立 Vectorization
from sklearn.svm import LinearSVC
vectorizer=CountVectorizer(tokenizer=tokenizer, ngram_range=(1,1), max_features=3800)
classifier=LinearSVC()
```

```python
x=pd.DataFrame(vectorizer.fit_transform(df['tweet']).toarray())
y=df['class']
```

```python
xtest=pd.DataFrame(vectorizer.fit_transform(dftest['tweet']).toarray())
#ytest=dftest['class']
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(criterion='gini', n_estimators=250, min_samples_split=350, oob_score=True,
                                 random_state=1, n_jobs=-1, )
classifier.fit(x,y)
```

```python
# 對測試集進行預測
ypred = classifier.predict(xtest)
print(len(ypred))
```

```python
row = list(dftest['id'])
d = {'id':row,'class':ypred}
test = pd.DataFrame(d)
test
```
